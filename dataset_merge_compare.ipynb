```python
import pandas as pd
import numpy as np

#   create 2 dummy datasets to test out logic
data = {
    'employee_id': range(1, 11),
    'first_name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie', 'Eve', 'David', 'Emily', 'Frank', 'Grace'],
    'last_name': ['Doe', 'Smih', 'Johnson', 'Williams', 'Brown', 'Taylor', 'Anderson', 'Davis', 'Lee', 'Jones'],
    'social': [f'{np.random.randint(100, 999)}-{np.random.randint(10, 99)}-{np.random.randint(1000, 9999)}' for _ in range(10)],
    'salary': np.random.randint(50000, 100000, size=10)
}

data1 = {
    'employee_id': range(1, 11),
    'first_name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie', 'Eve', 'David', 'Emily', 'Frank', 'Grace'],
    'last_name': ['Doe', 'Smih', 'Johnson', 'Williams', 'Brown', 'Taylor', 'Anderson', 'Davis', 'Lee', 'Jones'],
    'social': [f'{np.random.randint(100, 999)}-{np.random.randint(10, 99)}-{np.random.randint(1000, 9999)}' for _ in range(10)],
    'salary': np.random.randint(50000, 100000, size=10)
}
df = pd.DataFrame(data)
modified_df = pd.DataFrame(data1)

print(df)
print(f'\n{modified_df}')
```

```python
def print_changed(df1, df2, key_column):
    # Merge the original and modified DataFrames on the primary key
    merged_df = pd.merge(df1, df2, on=key_column, suffixes=('_old', '_new'))
    
    changed_values = []

    # Iterate through the rows
    for index, row in merged_df.iterrows():
        changes = {'emp_id': row[key_column]}
        
        # Compare values for each column pair
        for column in df1.columns:
            old_col = f'{column}_old'
            new_col = f'{column}_new'

            # Check if the old and new columns exist
            if old_col in merged_df.columns and new_col in merged_df.columns:
                old_value = row[old_col]
                new_value = row[new_col]

                if old_value != new_value:
                    changes[column] = {'old': old_value, 'new': new_value}
        
        if len(changes) > 1:  
            changed_values.append(changes)

    return changed_values
```

```python
key_column = 'employee_id'
```

```python
# dummy modification test
modified_df.at[2, 'salary'] = 9200
modified_df.at[3,'first_name'] = 'polamalu'
```

```python
print_changed(df, modified_df, key_column)
```

```python
keys = []
columns = []
old_values = []
new_values = []

# iterate through the result output
for entry in print_changed(df, modified_df, key_column):
    key = entry['emp_id']
    for column, values in entry.items():
        if column != 'emp_id':
            keys.append(key)
            columns.append(column)
            old_values.append(values['old'])
            new_values.append(values['new'])

# create the DataFrame
result_df = pd.DataFrame({
    'emp_id': keys,
    'column': columns,
    'old_value': old_values,
    'new_value': new_values
}).set_index('emp_id')
```
