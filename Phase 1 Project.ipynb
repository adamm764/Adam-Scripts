```python
product_db = [
    {'category id': 1, 'product id': 1, 'name':'Pea coat',     'price': 50, 'quantity' : 0},
    {'category id': 2, 'product id': 2, 'name': 'Fedora',      'price': 30, 'quantity': 0},
    {'category id': 2, 'product id': 3, 'name': 'Tophat',      'price': 25, 'quantity': 0},
    {'category id': 3, 'product id': 4, 'name':'Smartphone',   'price': 800, 'quantity': 0},
    {'category id': 3, 'product id': 5, 'name': 'iPad',        'price': 600, 'quantity': 0},
    {'category id': 4, 'product id': 6, 'name': 'Rain boots',  'price': 55, 'quantity': 0},
    {'category id': 4, 'product id': 7, 'name': 'Cowboy boots','price': 120, 'quantity': 0},
]
    
```

```python
categories_db = [
    {'category id': 1, 'name':'Jackets'},
    {'category id': 2, 'name':'Hats'},
    {'category id': 3, 'name':'Electronics'},
    {'category id': 4, 'name':'Boots'},
]
```

```python
def generate_session_id():
    return random.randint(100000, 999999)
```

```python
class AdminDB:
    
    def __init__(self, username, password):
        self.__username = username
        self.__password = password
        self.is_admin = True

    def authenticate_admin(self, username, password):
        if self.__username == username and self.__password  == password:
            return generate_session_id()
        print('Authentication failed')
        return 

#   Methods for Admins making changes to the product catalogue.
    def remove_product(self, product_id):
        for product in product_db:
            if product['product id'] == product_id:
                product_db.remove(product)
                print("product removed")
                return product_db

        print("Product not found.")
        return
    
    def add_product(self, category_id, product_id, name, price):
        for product in product_db:
            if product_id == product['product id']:
                print("Product already exists.")
                return
        new_product = {'category': category_id, 'product id': product_id , 'name': name, 'price': price, 'quantity' : 0}
        product_db.append(new_product)
        print("Product added.")
        return product_db
            
    def edit_product(self, product_id, new_category_id=None, new_name=None, new_price=None):
        if not self.is_admin:
            return "You do not have access to edit products."
 #      Using "= None", the admin can bypass updating values if they choose to do so 
        for product in product_db:
            if product['product id'] == product_id:
                if new_name is not None:
                    product['name'] = new_name
                
                if new_category_id is not None:
                    product['category id'] = new_category_id
                
                if new_price is not None:
                    product['price'] = new_price
                
                print("Product updated.")
                return product
        print("Product not found.")
        return
    
#   Methods for Admins making changes to the Category DB
    def add_category(self, new_category_id, new_category_name):
        for category in categories_db:
            if category['category id'] == new_category_id:
                print("Category already exists.")
                return
        new_category = {'category id': new_category_id, 'name': new_category_name}
        categories_db.append(new_category)
        print("Successfully added.")
        print(f"catalogue added: {new_category}")
        return categories_db
    
    def delete_category(self, category_id):
        for category in categories_db:
            if category['category id'] == category_id:
                categories_db.remove(category)
                print("Category deleted.")
                return categories_db
        print("Category not found.")
        return
            
```

```python
class UserDB:
    
    def __init__(self, username, password):
        self.__username = username
        self.__password = password
        self.user_cart = []
        self.is_admin = False
        
    def authenticate_user(self, username, password):
        if self.__username == username and self.__password == password:
            return generate_session_id()
        print("Authentication failed")
        return    
    
    def add_to_cart(session_id, user, product_id, quantity):
#       adding a counter to track the quantity parameter  
        item_quantity = 0
#       ensure that an admin is not trying to access user functions.     
        if user.is_admin:
            print("Access Denied. This function is reserved for users.")
            return

#       authenticate user, cut the method if user is not authenticated
        if user != session_id:
            print("Unable to verify user.")
            return
    
#       Find the dictionary for the product with the given product_id. NOTE : initiate a while loop to take in the quantity input
        for product in product_db:
            if product['product id'] == product_id:
                while item_quantity < quantity:

#       Adding a new product to the cart only if the item does not already exist, otherwise just adding it to the current quantity      
                    if product['quantity'] < 1:
                        user.user_cart.append(product)
                    else:
                        pass
                    print(f"1 {product['name']} added to cart.")
                    item_quantity += 1
                    product['quantity'] += 1
                return user.user_cart
#       If product_id not found in product_db, print an error message
        print("Product not found.")
        return user.user_cart
    

    def remove_from_cart(session_id, user, product_id, quantity):

#       this function is almost step for step the same as add_to_cart.     
        item_quantity = 0

        if user.is_admin:
            print("Access Denied, this function is reserved for users.")
            return

        if user != session_id:
            print("Unable to verify user.")
            return
    
        for product in user.user_cart:
            if product_id == product['product id']:
            
                while item_quantity < quantity:
                #   subtract from available quantity, if quantity no longer is greater than one, remove entire item.
                    if product['quantity'] > 1:
                        product['quantity'] -= 1
                    else:
                        user.user_cart.remove(product)

                    print(F"1 {product['name']} removed from cart.")
                    item_quantity += 1
            
                return user.user_cart 
                    
        print("Product not found")
        return user.user_cart
    
    
    def payment(self):
    
#       create empty string variable so it can be referenced before assignment    
        user_answer = ''
#       
        while user_answer.lower() not in ['mastercard', 'upi', 'paypal']:
        
            user_answer = input("Enter a payment option")
        
            if user_answer.lower() in ['mastercard', 'upi', 'paypal']:
                print("Thank you, you will be redirected to a new page to process payment.")
                return f"Payment selected: {user_answer}"

            else:
                print("Please enter a valid response.")
    

    def show_cart(session_id,user):
    
        total_price = 0
    
        if user.is_admin:
            print("Access Denied. Function reserved for users only.")
            return

        if user != session_id:
            return False
    
        if not user.user_cart:
            print("Cart empty.")
    
        else: 
            for item in user.user_cart:
                print(f"Your cart contains: {item['quantity']} {item['name']} totaling at {item['price'] * item['quantity']} US dollars.")
                total_price += item['price'] * item['quantity']        
#           Carry over the Payment method:
            return f"Your total price is: {total_price} USD", user.payment()
    
    
```
